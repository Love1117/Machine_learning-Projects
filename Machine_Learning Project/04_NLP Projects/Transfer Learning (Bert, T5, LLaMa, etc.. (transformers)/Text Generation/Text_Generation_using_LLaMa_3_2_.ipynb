##**Importing Libraries and connecting my coding environment to Hugging face**

```python
import os
from openai import OpenAI

# Set my Hugging Face API token.
# IMPORTANT: For production, consider using Colab's 'Secrets' feature for better security.
# Example: from google.colab import userdata
# os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN') # Assuming you've saved your token as 'HF_TOKEN' in Colab Secrets.
os.environ["HF_TOKEN"] = "Love1117" # Replace "YOUR_HF_TOKEN" with your actual token

client = OpenAI(
    base_url="https://router.huggingface.co/v1",
            api_key=os.environ["HF_TOKEN"],
                    )
```

##**Creating a function which i can reprosticate a chatbot environment type.**

```python
def chat_bot():
  user_input = input("I am your Ai chatboot, you can ask me any question (Ask or Quit): ")
  while True:
    if user_input.lower() == "ask":
      query_text = input("Please kindly ask your question or (stop):  ")
      if query_text == "stop":
        print("It a wonderful experience with you. see you next time")
        break

      else:
        completion = client.chat.completions.create(
              model="meta-llama/Llama-3.2-3B-Instruct:novita",
                          messages=[{"role": "user",
                                    "content": query_text}])
        print(completion.choices[0].message.content)
        print("\n")
        print("\n")
    elif user_input.lower() == "quit":
      print("Alright, have a great day!")
      break
    else:
      print("I didn't understand that. Please type 'Ask' or 'Quit'.")
      break
```
