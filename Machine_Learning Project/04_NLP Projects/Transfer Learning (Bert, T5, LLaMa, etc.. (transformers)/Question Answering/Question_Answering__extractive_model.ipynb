##**writing a function that will select our starting index and ending index then decode my answer**

```python
def answer_question(output_idx, input_idx, question_text_idx):
  answer_start_idx = torch.argmax(output_idx.start_logits)
  answer_end_idx = torch.argmax(output_idx.end_logits)

  answer_token = input_idx.input_ids[0, answer_start_idx : answer_end_idx + 1]
  answer = tokenizer.decode(answer_token)

  print(f"""Question: {question_text_idx['question']}
  Answer: {answer}""")
```

#**Method 2:** Using Transformer Pipeline

```python
from transformers import pipeline
pipeline = pipeline("question-answering", model = model_link, tokenizer=model_link)
```

```python
def outputs(question_text_idx):
  output = pipeline(question_text_idx["question"], question_text_idx["context"])
  return output
```
