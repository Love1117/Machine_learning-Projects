##**Installation and importing of libraries for my project**

```python
!pip install pytextrank
import spacy
import pytextrank
```

##**Load spaCy model and Add Text rank into pipeline**

```python

# Load spaCy model
model = spacy.load("en_core_web_sm")

# Add TextRank into spaCy pipeline
model.add_pipe("textrank")
```

##**Input Text**

```python

text = """Loveday Shadrack had always been the kind of person who carried two worlds inside him: one filled with code, logic, and problem-solving and the other filled with rhythm, melody, and the voice of a storyteller. From the dusty corners of his childhood notebooks to the glowing screen of his first computer, he discovered early that he didnt just love learninghe needed it.

He stopped seeing coding as just a skill. It became a calling. He stopped seeing singing as a hobby. elf.

Slowly, he began to realize that greatness wasnt a momentit was a journey. And he was walking it faithfully.

Loveday Shadrack, once just a boy with curiosity, was becoming a young man with purpose. A man sharpening his mind, strengthening his spirit, refining his talents, and pushing forward even when life felt heavy.

His story is still unfolding, but one thing is already clear:

He is risingand nothing, not hardship, not doubt, not fear, can stop the man God is shaping him to be."""
```

##**Process text**

```python
doc = model(text)
```

##**Summarizing Text: printing top sentences**

```python
print("---- SUMMARY ----")
for sent in doc._.textrank.summary(limit_phrases=5, limit_sentences=6):
    print(sent)
```

##**Adjustable parameters such as limit_phrases and limit_sentences were used to control the depth, size, and focus of the generated summaries, making the system flexible for different use cases.**
